## Pregunta 1

Se quería analizar la distribución de los puntajes tanto de IMDB como de Rotten Tomatoes para las películas de Marvel y DC. Para lo anterior, se crearon 4 histogramas, en el eje x se distribuyen los puntajes de IMDB Rating de Rotten Tomatoes por separado, para cada entidad.
Los pasos con los que se hicieron son los siguientes:

1.	Se creo un filtro del dataframe para seleccionar la entidad que se quiere
2.	Se filtran los datos del dataframe a un subset  a partir del filtro
3.	Se establece que el grafico a realizar es una figura y le ponemos el tamaño deseado
4.	Con el método de la librería matplotlib hist() se crea el histograma
5.	Con el método kde (que significa “kernel distribution estimation”) se le agrega al grafico la línea de densidad, ademas se le agrego la variable ind que se le asigno un array (para esto se importo NumPy al principio) con todos los numeros a evaluar desde el valor minimo al maximo
6.	Por ultimo se le agregan nombres a los ejes y titulo al grafico, con plt.xlabel, plt.ylabel y plt.title.

Referencias utilizadas:
* *https://www.youtube.com/watch?v=QPaHUBphDgo*
* *https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.kde.html*
* *https://www.skytowner.com/explore/normalizing_a_histogram_in_matplotlib*




````python
filt = (df["entity"] == "DC") #se le aplica el filtro: la entidad sea DC
data_0 = df.loc[filt, "imdb_rating"] #filtramos los datos de los rating IMDB a partir del filtro
plt.figure(figsize=(8, 6)) #establecemos el tamano del grafico

imdb_dc = plt.hist(data_0, bins="auto", density=True, color="cornflowerblue", edgecolor="black") #histograma del df
data_0.plot.kde(color="orange", linestyle="--", ind=np.arange(min(data_0), max(data_0), 0.1))#linea de densidad
plt.xlabel("IMDB Rating")
plt.ylabel("Probabilidad")
plt.title("Histograma IMDB Rating de DC")

plt.show()
````

El eje x es el puntaje en IMDB, mientras mas a la derecha mas alto es este, y el eje y es la probabilidad de densidad del total de peliculas de DC.

De este grafico se puede analizar que las películas de DC tienen mucha probabilidad de estar en un rating de entre 5 y 7 de IMDB, ya que alrededor del 25% se encuentra ahí.

Por último, se unieron los 4 gráficos en una misma figura mediante subplots. Para esto:

4.	Los gráficos se hicieron de la misma manera que antes
5.	Para finalizar se utilizo .tight_layout() en la figura completa, para ajustar de mejor manera los espacios

Referencias:
* *https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html*
* *https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html*
* *https://www.statology.org/matplotlib-subplot-spacing/*

````python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle('Histogramas Rating Marvel-DC\n', fontsize=30)
datos = [data_0, data_1, data_2, data_3]
labels = ["IMDB Rating", "Tomato Meter"]
entidad = ["DC", "Marvel"]
pos = 0

for i in range(2):
    for t in range(2):
        plt.subplot(2, 2, pos+1)
        plt.hist(datos[pos], bins="auto", density=True, color="cornflowerblue", edgecolor="black")
        datos[pos].plot.kde(color="orange", linestyle="--", ind=np.arange(min(datos[pos]), max(datos[pos]), 0.1))
        ax[i, t].set_ylabel('Probabilidad')
        ax[i, t].set_xlabel(labels[i])
        ax[i, t].set_title(f"Histograma {labels[t]} de {entidad[t]}\n")
            
        pos += 1

fig.tight_layout()
plt.show()
````

Con todos los gráficos al mismo tiempo se puede concluir que Marvel tiene mayor tendencia a obtener mejores puntajes de rating tanto en IMDB como en el Tomato Meter que DC.

# Pregunta 2

Para esta pregunta buscamos realizar un gráfico de dispersión de películas según su puntaje. En el eje x, estan los puntajes según IMDB, y en el eje y estan los puntajes de Rotten Tomatoes. Además, el tamaño de los puntos corresponde a la variable Runtime de cada película y por último, el color corresponde a la variable “Mpa Rating”.
Para la confección de este grafico se hizo lo siguiente:
1.	Se crearon filtros según su mpa rating
2.	Se almacenaron en una lista, y se estableció otra lista con sus colores correspondientes
3.	Luego se estableció la figura y mediante el método de matplotlib .scatter() se logro hacer el grafico indicado.
4.	Se hicieron 5 ciclos para ingresar las distintas series de datos automáticamente, indicando en una lista los runtimes y asignadosela a la variable *size* de scatter.

Referencias:
* *https://www.youtube.com/watch?v=zZZ_RCwp49g*
* *https://github.com/pandas-dev/pandas/issues/16827*
* *https://stackoverflow.com/questions/4270301/matplotlib-multiple-datasets-on-the-same-scatter-plot*
* *https://www.geeksforgeeks.org/how-to-increase-the-size-of-scatter-points-in-matplotlib/*
* *https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html*


````python
mpa_approved = (df["mpa_rating"] == "Approved")
mpa_pg = (df["mpa_rating"] == "PG")
mpa_pg13 = (df["mpa_rating"] == "PG-13")
mpa_R = (df["mpa_rating"] == "R")
mpa_passed = (df["mpa_rating"] == "Passed")

filtros = [[mpa_approved, "Approved"], [mpa_passed,"Passed"], [mpa_pg, "PG"], [mpa_pg13, "PG-13"], [mpa_R, "R"]]
colors = ['g', 'r', 'b', 'c', 'm']
````

````python
plt.figure(figsize=(8, 6))

for i in range(5):

    x = df.loc[filtros[i][0], "imdb_rating"]#eje x
    y = df.loc[filtros[i][0], "tomato_meter"]#eje y
    size = df.loc[filtros[i][0],  "runtime"]#runtimes
    color = colors[i]#color de la serie

    plt.scatter(x, y, s=size, c=color, alpha=0.3, label=filtros[i][1])

################################################################################################
#Esta parte del codigo se saco de la referencia 2 de arriba, sirve para hacer la leyenda
#del runtime por porte de los puntos
pws = (pd.cut(df["runtime"], bins=5, retbins=True)[1]).round(0)
for pw in pws:
    plt.scatter([], [], s=(pw**2)/2e4, c="k",label=int(pw))
    
h, l = plt.gca().get_legend_handles_labels()
plt.legend(h[5:], l[5:], labelspacing=1.2, title="Runtime", borderpad=1, 
            frameon=True, framealpha=0.9, numpoints=1, markerscale=8, bbox_to_anchor = (1 , 1))
#################################################################################################

plt.xlabel("IMDB Rating")
plt.ylabel("Tomato Meter")
plt.title("IMDB Rating v/s Tomato Meter\n", fontsize=20)
plt.show()
````

El eje x es el puntaje de una pelicula en IMDB, y el eje y es el puntaje en Rotten Tomatoes, por lo que si un punto esta ubicado mas arriba a la derecha tiene mejor puntacion en ambas. Ademas el punto tiene un porte determinado por su runtime y su color determinado por su MPA Rating.

De este grafico se puede ver que la mayoría de las películas de Marvel y DC están calificadas con MPA Rating de PG-13 y tienen un Runtime similar. Además se puede ver que el MPA Rating de una película no tiene relación con su calificación en IMDB o Tomato Meter. Finalmente que los MPA Rating "Passed" y "Approved" fueron excepciones que recibieron mal puntaje en Tomato Meter.

# Pregunta 4

Para la confeccion de este grafico se utilizo la libreria seaborn, especificamente la funcion regplot. Se hicieron distintos datasets por año, y sus puntajes en IMDB y Rotten Tomatoes.

Referencias:
* *https://www.geeksforgeeks.org/python-seaborn-regplot-method/*
* *https://seaborn.pydata.org/generated/seaborn.regplot.html*

````python
data_6 = df[(df["year"]>=2005) & (df["year"]<=2015)]

data_7 = data_6.loc[filt_marvel, ["year", "imdb_rating"]]
data_8 = data_6.loc[filt_marvel, ["year", "tomato_meter"]]
data_9 = data_6.loc[filt_dc, ["year", "imdb_rating"]]
data_10 = data_6.loc[filt_dc, ["year", "tomato_meter"]]
````

````python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle('Ratings Promedio por año\n', fontsize=30)

plt.subplot(2, 2, 1)
sns.regplot(x = "year", y = "imdb_rating", data = data_7, x_estimator=np.mean)
ax[0, 0].set_ylabel("IMDB Rating")
ax[0, 0].set_xlabel("Año")
ax[0, 0].set_title("IMDB Rating de DC\n")

plt.subplot(2, 2, 2)
sns.regplot(x = "year", y = "tomato_meter", data = data_8, x_estimator=np.mean)
ax[0, 0].set_ylabel("Tomato Meter")
ax[0, 0].set_xlabel("Año")
ax[0, 0].set_title("Tomato Meter de DC\n")


plt.subplot(2, 2, 3)
sns.regplot(x = "year", y = "imdb_rating", data = data_9, x_estimator=np.mean)
ax[0, 0].set_ylabel("IMDB Rating")
ax[0, 0].set_xlabel("Año")
ax[0, 0].set_title("IMDB Rating de DC\n")


plt.subplot(2, 2, 4)
sns.regplot(x = "year", y = "tomato_meter", data = data_10, x_estimator=np.mean)
ax[0, 0].set_ylabel("Tomato Meter")
ax[0, 0].set_xlabel("Año")
ax[0, 0].set_title("Tomato Meter de DC\n")


fig.tight_layout()
plt.show()
````
